```python
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def power_law(wealth):
    wealth += max(np.random.uniform(.0, .2) * wealth, 0)
    return wealth + np.random.normal()
    
def normal(wealth):
    return wealth + np.random.normal()

def wealth_after_60_yrs(wealth_fn):
    wealth = wealth_fn(0)
    for _ in range(60):
        wealth = wealth_fn(wealth)
        
    return wealth
```

```python
sns.distplot([wealth_after_60_yrs(normal) for _ in range(9999)])
```

```python
sns.distplot([wealth_after_60_yrs(power_law) for _ in range(9999)])
```
